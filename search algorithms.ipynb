```python
def goal_test(node, goal=Goal):
    if node == goal:
        return True
    else:
        return False
```

```python
def tupleConversion(node):
    node_tuple = []
    for sublist in node:
        node_tuple.append(tuple(sublist))
    return tuple(node_tuple)
```

```python
def find(node):
    for i in range(3):
        for j in range(3):
            if node[i][j] == 0:
                return i, j
            
```

```python



def bfs(source_node, goal=Goal):
    frontier = queue.Queue()
    frontier.put((-1,0,source_node))
    source_tuple = tupleConversion(source_node)
    explored_set = set()
    prev_node = [(-1,0,source_node)]
   
    goal_reached = False
    
    
    while not frontier.empty():
        
        
        p_node = frontier.get()
        node=p_node[2]
        node_tuple = tupleConversion(node)
        if goal_test(node, goal):
            goal_reached = True
            break
        else:
            if  node[0][0] != 0 and node[1][0] != 0 and node[2][0] != 0:
                child = deepcopy(node)
                i, j = find(child)
                child[i][j],child[i][j-1]=child[i][j-1],child[i][j]
                child_tuple = tupleConversion(child)
                if child_tuple not in explored_set:
                    if goal_test(child,goal):
                        goal_reached=True
                        prev_node.append((p_node[1],len(prev_node),child))
                        break
                    else:
                        frontier.put((p_node[1],len(prev_node),child))
                        prev_node.append((p_node[1],len(prev_node),child))
                    
                    

            if  node[0][2] != 0 and node[1][2] != 0 and node[2][2] != 0:
                child = deepcopy(node)
                i, j = find(child)
                child[i][j],child[i][j+1]=child[i][j+1],child[i][j]
                child_tuple = tupleConversion(child)
                if child_tuple not in explored_set:
                    if goal_test(child,goal):
                        goal_reached=True
                        prev_node.append((p_node[1],len(prev_node),child))
                        break
                    else:
                        frontier.put((p_node[1],len(prev_node),child))
                        prev_node.append((p_node[1],len(prev_node),child))
                    
            if  node[0][0] != 0 and node[0][1] != 0 and node[0][2] != 0:
                child = deepcopy(node)
                i, j = find(child)
                child[i][j],child[i-1][j]=child[i-1][j],child[i][j]
                child_tuple = tupleConversion(child)
                if child_tuple not in explored_set:
                    if goal_test(child,goal):
                        goal_reached=True
                        prev_node.append((p_node[1],len(prev_node),child))
                        break
                    else:
                        frontier.put((p_node[1],len(prev_node),child))
                        prev_node.append((p_node[1],len(prev_node),child))
                    

            if  node[2][0] != 0 and node[2][1] != 0 and node[2][2] != 0:
                child = deepcopy(node)
                i, j = find(child)
                child[i][j],child[i+1][j]=child[i+1][j],child[i][j]
                child_tuple = tupleConversion(child)
                if child_tuple not in explored_set:
                    if goal_test(child,goal):
                        goal_reached=True
                        prev_node.append((p_node[1],len(prev_node),child))
                        break
                    else:
                        frontier.put((p_node[1],len(prev_node),child))
                        prev_node.append((p_node[1],len(prev_node),child))
                    
        explored_set.add(node_tuple)
    return prev_node   
        
    
```

```python
def Heuristic(node, goal):
    sum = 0
    for i in range(3):
        for j in range(3):
            value = node[i][j]
            for k in range(3):
                for l in range(3):
                    if value == goal[k][l]:
                        sum += abs(i-k) + abs(j-l)

    return sum
```
